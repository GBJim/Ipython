```python
# Let's visualize the survial rate between gender


def genderBar(data):   # Define a function which plots Barchart
    x = pd.Categorical.from_array(data['Sex'])  # Categorize the string-typed Sex column
    x=x.levels           
    
    male=data[data.Sex=='male']                 # Select Male and Female data rows
    female=data[data.Sex=='female']
    
    numberOfMale=len(female)                   # Compute the population of male and female
    numberOfFemale=len(male)
 
    male=male[male.Survived==1].count()/male.count()            #Compute the survival rate of gender
    female=female[female.Survived==1].count()/female.count()
    y=(female.Survived,male.Survived)
    # Start to plot
    plt.ylim(0,1)                                                                       # Set Y axis range 
    plt.xlabel('Female:'+str( numberOfFemale)+'   Male:'+str( numberOfMale))              # Label of X axis 
    plt.xticks(np.arange(len(x)), x)                                                    # Set ticks of X axis
    plt.bar(np.arange(len( x)),y,align='center',color='blue',width=0.35)                # Deploy the barChart

plt.title("Total  "+str(len(titanic))+'people')      # Set the title                   
genderBar(titanic)
plt.show()                      # Show the plot
```

```python
# In this section, we group the data by Pclass and plot a barchart for each group

PclassGroup=titanic.groupby(['Pclass'])
plt.figure(1)
i=0
for name, group in PclassGroup:
    #print name
    #print group.describe()
    
    i+=1
    plt.subplot(130+i)
    plt.title("Pclass"+str(name)+'  ' +str(len(group))+'people')
    
    genderBar(group)
    
plt.ylim(0,1)
plt.tight_layout()  
plt.show()    


    
    
    
    
    
    
```
